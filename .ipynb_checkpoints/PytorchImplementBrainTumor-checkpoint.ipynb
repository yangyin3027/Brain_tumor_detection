### Study Overiew
- The image datasets are labeled with tumor yes and no. A model will be learnt to classify the images with tumor yes or no. It could facilitate the radiological intepratation of diagnostic images.
- However, this is a small datasets for sophisticated CNN or other deep neural network learning. A number of strategies are implemented to mitigate the size limitation of the datasets, included:
    - Image augmentation by adding random rotation, width/height shift, random shear, brightness, horizontal/vertical flip to increase volume of the datasets.
    - Explotatory data analysis showed that the brain images were cropped from larger whole body image with varied black border, which doesn't contribute to the learning. To avoid waste of computing resources and overfitting, images were preprocessed using `findContour` function in `OpenCV-Python` to find the largest contour and thus define the bounding box of each image. Each image was cropped using the refined bounding box to remove the black border.
- <font size='4'>**Main experimentation performed in this study:**</font>
    - <font size='3'>Tried `Pytorch` implemented a shallow `CNN` model, and render it to `sklearn estimator` object as well as later to a customized `calibration` object for `Classifier Calibration` task.</font>
    - <font size='3'>Fine tuned couple of `sklearn classifiers`.</font>
    - <font size='3'>Performed `Classifier Calibration` for the trained classifiers.</font>

#### 2.2 Pytorch-based Deep Neural Networks

- <font size='3'>Implement a pytorch CNN models and render it a `sklearn base estimator` so that downstream model performance evaluation generated on general sklearn base estimator could be applicable to the pytorchCNN as well.</font>

```python
class CNNClassifier(nn.Module):
    def __init__(self, classes=2):
        super().__init__()
        
        self.features = nn.Sequential(
                            nn.Conv2d(3, 32, 3, 1, 1),
                            nn.ReLU(),
                            nn.BatchNorm2d(32),
                            nn.MaxPool2d((2,2)),)
        
        self.avgpool = nn.AdaptiveAvgPool2d((5,5))
        
        self.classifier = nn.Sequential(
                            nn.Linear(25*32, 128),
                            nn.ReLU(inplace=True),
                            nn.Dropout(0.5, inplace=False),
                            nn.Linear(128, classes),
                            nn.Sigmoid())
        
        self._init_weights()
        
        self.optim = torch.optim.RMSprop(self.parameters(), lr=0.001)
        self.criterion = nn.BCELoss()
        
        self.train_loss = []
        self.valid_loss = []
    
    def forward(self, x):
        x = self.features(x)
        x = self.avgpool(x)
        x = x.reshape(x.size(0), -1)  # avg_pool forward 
        y = self.classifier(x)
        return y
    
    def _init_weights(self):
        for m in self.modules():
            if isinstance(m, nn.Conv2d):
                torch.nn.init.xavier_normal(m.weight)
                m.bias.data.fill_(0.01)
            if isinstance(m, nn.Linear):
                torch.nn.init.xavier_uniform(m.weight)
                m.bias.data.fill_(0.01)
                
    def train_with_dataloader(self,dataloader, valid_dataloader, num_epochs=20, plot=False):
            
        for epoch in range(num_epochs):
            
            running_loss = 0.0
            for X, y in dataloader:
                loss = self._single_train(X, y)
                running_loss += loss
            
            epoch_loss = running_loss / len(dataloader.dataset)
            self.train_loss.append(epoch_loss)
            
            running_valid_loss = 0.0
            with torch.no_grad():
                for X, y in valid_dataloader:
                    yhat = self(X)[:, 1]
                    valid_loss = self.criterion(yhat, y.float())
                    running_valid_loss += valid_loss.item() * X.size(0)
                self.valid_loss.append(running_valid_loss/len(valid_dataloader.dataset))
            
            if plot:
                plt.gca().cla()
                plt.plot(np.arange(1, len(self.train_loss)+1), self.train_loss, 
                         c='m', label='Training Loss')
                plt.plot(np.arange(1, len(self.valid_loss)+1), self.valid_loss, 
                         c='darkorange', label='Valid Loss')
                plt.xlabel('Epoch Number')
                plt.ylabel('BCELoss')
                plt.grid()
                plt.legend()
                display.clear_output(wait=True)
                display.display(plt.gcf())
        return self
            
    
    def _single_train(self, X, y):
        yhat = self(X)[:, 1]
        loss = self.criterion(yhat, y.float())
        loss_val = loss.item() * X.size(0)
        
        self.optim.zero_grad()
        loss.backward()
        self.optim.step()
        return loss_val
        
    
    def train(self, X, y, batch_size=32, num_epochs=20, verbose=False):
        
        for epoch in range(num_epochs):
            running_loss = 0.0
            for X_, y_ in zip(*[torch.split(i,batch_size) for i in (X, y)]):
                loss = self._single_train(X_, y_)
                running_loss += loss
            
            epoch_loss = running_loss / len(y)
            if verbose:
                print(f'Epoch-{epoch} ->>> Loss: {epoch_loss:.3f}')
        return self
        
# instantiate model and load datasets

cnn_clf = CNNClassifier()

train_dataset = TumorDataset(X_train, y_train, transform(train=True))
train_dataloader = get_dataloader(train_dataset)

valid_dataset = TumorDataset(X_valid, y_valid, transform(train=False))
valid_dataloader = get_dataloader(valid_dataset)
```

#### 2.3 Render the `Pytorch CNNClassifier` a `sklearn base estimator` object. 

- <font size='3'>Make it compartible to downstream `sklean` packages for `model evaluation` and `classifier calibration` </font>

```python
class TorchModel(ScikitClassifier):
    def __init__(self, model):
        self.model = model
    
    def fit(self, X, y, verbose=False):
        X = torch.stack([transform(True)(img) for img in X])
        y = torch.Tensor(y)
        return self.model.train(X, y, batch_size=32, num_epochs=20, verbose=verbose)
        
    def predict(self, X):
        if X.ndim < 3:
            X = [X]
        X = torch.stack([transform(False)(img) for img in X])
        with torch.no_grad():
            yhat = self.model(X)
            _, pred = torch.max(yhat, dim=-1)
        return pred
        
    def predict_proba(self, X):
        if X.ndim < 3:
            X = [X]
        X = torch.stack([transform(False)(img) for img in X])
        with torch.no_grad():
            yhat = self.model(X)
        return yhat[:, 1]

cnn_clf_sk = TorchModel(cnn_clf)
```

- <font size='3'> Customize a `Classifier Calibration` class for `Pytorch model` </font>

```python
class SigmoidCalibrator:
    def __init__(self, prob_pred, prob_true):
        prob_pred, prob_true = self._filter_out_of_domain(prob_pred, prob_true)
        prob_true = np.log(prob_true/(1-prob_true)) # transforming the datapoints to ln(x/(1-x))
        
        # use linearRegressor to make prob_pred
        self.regressor = LinearRegression().fit(prob_pred.reshape(-1, 1), prob_true.reshape(-1, 1))
    
    def calibrate(self, probabilities):
        # use sigmoid function (1 /(1+exp(-x))) to tranform datapoints back to original domain
        return 1/(1+np.exp(-self.regressor.predict(probabilities.reshape(-1,1)).flatten()))
    
    def _filter_out_of_domain(self, prob_pred, prob_true):
        filtered = list(zip(*[p for p in zip(prob_pred, prob_true) if 0 < p[1] < 1]))
        return np.array(filtered)

class IsotonicCalibrator:
    def __init__(self, prob_pred, prob_true):
        self.regressor = IsotonicRegression(out_of_bounds='clip')
        self.regressor.fit(prob_pred, prob_true)
        
    def calibrate(self, probabilities):
        return self.regressor.predict(probabilities)

class CalibratableModelMixin:
    def __init__(self, model):
        self.model = model
        self.name = model.__class__.__name__
        self.sigmoid_calibrator = None
        self.isotonic_calibrator = None
        self.calibrators = {
            'sigmoid': None,
            'isotonic': None,
        }
    
    def calibrate(self, X, y):
        predictions = self.predict(X)
        prob_true, prob_pred = calibration_curve(y, predictions, n_bins=10)
        self.calibrators['sigmoid'] = SigmoidCalibrator(prob_pred, prob_true)
        self.calibrators['isotonic'] = IsotonicCalibrator(prob_pred, prob_true)
    
    def calibrate_probabilities(self, probabilities, method='isotonic'):
        if method not in self.calibrators:
            raise ValueError("Methods has to be either 'sigmoid' or 'isotonic'")
        if self.calibrators[method] is None:
            raise ValueError('Run calibrate to fit the calibrators first')
        return self.calibrators[method].calibrate(probabilities)
    
    def predict_calibrated(self, X, method='isotonic'):
        return self.calibrate_probabilities(self.predict(X), method)
    
    def score(self, X, y):
        return self._get_accuracy(y, self.predict(X))
    
    def score_calibrated(self, X, y, method='isotonic'):
        return self._get_accuracy(y, self.predict_calibrated(X, method))
    
    def _get_accuracy(self, y, preds):
        return np.mean(np.equal(y.astype(bool), preds >= 0.5))

class NNModel(CalibratableModelMixin):
    def fit(self, X, y):
        self.model.fit(X, y)
    
    def predict(self, X):
        return self.model.predict_proba(X)

cal_cnn_clf_sk = NNModel(cnn_clf_sk)
setattr(cnn_clf_sk, 'name', 'cnn_clf')
setattr(cal_cnn_clf_sk, 'name', 'cnn_clf_calibrated')
```

#### 2.4 Implement conventional machine learning models

```python
def train_classifiers(models, params, X, y):
    trained_models = []
    
    for model, param in zip(models, params):

        rf_cv = RandomizedSearchCV(estimator=model,
                                  param_distributions=param,
                                  n_iter=100,
                                  cv=3,
                                  verbose=2,
                                  random_state=42,
                                  n_jobs=-1)

        rf_cv.fit(X, y)
        
        trained_models.append(rf_cv.best_estimator_)
        
    return trained_models

models = [RandomForestClassifier(),
         SVC()]
params = [dict(n_estimators = [int(x) for x in np.linspace(200,2000,10)],
            max_depth = [int(x) for x in np.linspace(3,10,3)] + [None],
            min_samples_split = [2,5,10],
            min_samples_leaf = [1,2,4]),
          
         dict(C=np.linspace(0.01, 100, 5))]

# preprocess training_dataset
scaler = StandardScaler()
X_train_flatten = X_train.reshape(X_train.shape[0], -1)
X_train_scaled = scaler.fit_transform(X_train_flatten)

X_valid_flatten = X_valid.reshape(X_valid.shape[0], -1)
X_valid_scaled = scaler.transform(X_valid_flatten)

X_val_flatten = X_val.reshape(X_val.shape[0], -1)
X_val_scaled = scaler.transform(X_val_flatten)

trained_models = train_classifiers(models, params, X_train_scaled, y_train)
```

### 3. Model Performance Evaluation and Classifier Calibration

#### 3.1 Pack Calibrated and Uncalibrated Models

```python
def calibrate_models(models, X, y):
    new_models = []
    
    # performaning classifier calibration
    for model in models:
        setattr(model, 'name', model.__class__.__name__)
        new_models.append(model)
        cal_model = CalibratedClassifierCV(model, cv=3)
        cal_model.fit(X, y)
        setattr(cal_model, 'name', f'{model.__class__.__name__}_calibrated')
        new_models.append(cal_model)
    return new_models

calibrated_models = calibrate_models(trained_models, X_train_scaled, y_train)
```

#### 3.2 Visualize the performance and calibration

```python
def plot_calibration_details(models, X, X_scaled, y):
    '''
    Args:
        X: for cnn training
        X_scaled: flatten and scaled for sklearn training
    '''
    
    plt.figure(figsize=(12, 8),layout='constrained')
    ax1 = plt.subplot2grid((5,1), (0,0), rowspan=2)
    ax2 = plt.subplot2grid((5,1), (2,0), rowspan=1)
    ax3 = plt.subplot2grid((5,1), (3,0), rowspan=2)
    
    for model in models:
        name = model.name
        if not name.endswith('_calibrated'):
            if name.startswith('cnn_clf'):
                pred_proba = model.predict_proba(X)
            elif name.startswith('SVC'):
                continue
            else:
                pred_proba = model.predict_proba(X_scaled)[:, 1]
        else:
            if name.startswith('cnn_clf'):
                pred_proba = model.predict(X)
            else:
                pred_proba = model.predict_proba(X_scaled)[:, 1]
        
        prob_true, prob_pred = calibration_curve(y, pred_proba, n_bins=10)
        brier_score = brier_score_loss(y, pred_proba)
        
        ax1.plot(
            prob_pred,
            prob_true,
            marker='.',
            label=f'{name}(brier_score={brier_score:.3f})'
            )
        
        ax2.hist(
            pred_proba, range=(0,1), bins=10, label=name, histtype='step', lw=2)
        
        RocCurveDisplay.from_predictions(
            y,
            pred_proba,
            name=name,
            ax=ax3)
    ax1.plot([0, 1], [0, 1], ls='--',label='Perfectly cablirated' )
    ax1.set_ylabel("Fraction of positives")
    ax1.set_ylim([-0.05, 1.05])
    ax1.legend(loc="lower right")
    ax1.set_title("Calibration plots")
    ax1.grid()

    ax2.set_xlabel("Mean predicted value")
    ax2.set_ylabel("Count")
    ax2.legend(loc="upper left", ncol=2)
    ax2.grid()
    
    ax3.plot([0, 1], [0, 1], ls='--',label='Chance level (AUC = 0.5)' )
    ax3.set_xlabel('False Positive Rate')
    ax3.set_ylabel('True Positive Rate')
    ax3.set_title('AUC')

all_models = calibrated_models + [cnn_clf_sk, cal_cnn_clf_sk]
plot_calibration_details(all_models, X_val, X_val_scaled, y_val);
```

- <font size='4'>**Notes:** Classifier calibration on `cnn` model is not necessary and did not improve its `brier_score_loss` or `AUC`, which was anticipated as `sigmoid` layer was already included in the `cnn` model. As for other models, naive `SVC` model does not provide probability prediction, therefore, classifier calibration is very helpful.</font>

- <font size='4'> Overall, the biggest problem in this study is the small datasets. So the `Classifiers` are still heavily deviated from perfectly calibrated curve.</font>
