## 2. Reorganize Dataset and Split to Train/Validation/Test Set

```python
import os
import random
from shutil import copyfile
from pathlib import Path
from tqdm import tqdm

def copy_files(source, destination, files):
    name = destination.split('/')[-2:]
    for file in tqdm(files, total = len(files), 
                     desc = f'Generating and copying files to {name}'):
        thisfile = os.path.join(source, file)
        targetfile = os.path.join(destination, file)
        copyfile(thisfile, targetfile)

def split_train_val_test(source_dir, dir_root ='', test_num = 5, split_ratio = 0.8):
    '''
    Traverse subfolders that represent each category and split it to train/test/valid datasets.
    New organized train/test/valid structured folders are created.
    Data files in each subfolders are copied to designated structured folders
    
    dir_root: as default it is the current working directory
    test_num: it is defaulted as 5, could be changed depending on the size of the datasets
    split_ratio: it is regarded as train/valid ratio
    '''
    # get collections of classes subfolders
    class_folders = [x for x in os.listdir(os.path.join(dir_root, source_dir))
                    if not x.startswith('.')]
    # traverse each class_folders, perform split on each class
    for cat in class_folders:
        # create 'train'&'valid' for each class folder
        
        print(f"Splitting {cat} Dataset")
        source_folder = os.path.join(dir_root, source_dir, cat)
        train_folder = os.path.join(dir_root,'datasets', 'train', cat)
        test_folder = os.path.join(dir_root, 'datasets', 'test', cat)
        valid_folder = os.path.join(dir_root,'datasets','valid', cat)
        
        
        Path(train_folder).mkdir(parents = True, exist_ok = True)
        Path(test_folder).mkdir(parents = True, exist_ok = True)
        Path(valid_folder).mkdir(parents = True, exist_ok = True)
        
        files = [x for x in os.listdir(os.path.join(dir_root, source_dir, cat))
                if not x.startswith('.')] # in case unexpected system files exist
        
        # get train/test/valid subset
        shuffle_set = random.sample(files, len(files))
        train_num = int(len(files)*split_ratio)
        
        test_set = shuffle_set[:test_num]
        train_set = shuffle_set[test_num:train_num]
        valid_set = shuffle_set[train_num:]
        
        # perform copy files
        copy_files(source_folder, train_folder, train_set)  
        copy_files(source_folder, test_folder, test_set)
        copy_files(source_folder, valid_folder, valid_set)
```

## 3. Data Import and Preprocessing
Image data is loaded as np.array. Multiple `imread` method could be used, for example, `cv2.imread`, `skimage.io.imread` and `scipy.misc.imread`. Be noted, the defaulted channel sequence in `cv2.imread` is `BGR`, which should be transformed using `cv2.cvtColor(src, cv2.COLOR_BGR2RGB)` before it could be natively displayed using `plt.imshow`.
<br>
<br>
The category of the dataset is binary as `Yes` and `No` tumor type. They will be encoded as `1` and `0` for labels using a dictionary.

### 3.1 Loading data

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


def load_data(data_dir):
    '''
    Load image files to numpy.array data and encode each class to a label dictionary.
    Subfolder inside the data_dir represent each class
    '''
    X = []
    y = []
    labels = dict()
    i = 0
    
    for path in tqdm(sorted(os.listdir(data_dir))):
        if not path.startswith('.'):  # to avoid system folders
            labels[i] = path
            
            for file in os.listdir(os.path.join(data_dir, path)):
                if not file.startswith('.'): # to avoid system files
                    fname = os.path.join(data_dir, path, file)
                    img = cv2.imread(fname)
                    X.append(img)
                    # path represents each class and now be encoded to categorical number i
                    y.append(i)
            i += 1
    X = np.array(X, dtype = object)
    y = np.array(y, dtype = object)
    
    print('\n',f'{len(X)} images loaded from {data_dir} directory.')
    return X, y, labels

def enlarge_with_padding(src, img_size, padding):
    '''
    Enlarge the source image to a bigger size with padding values
    src: it should be np.array representation of image 
    
    img_size: it is a image shape tuple with dimensions larger than the 
    orginal image
    
    padding: it needs to match with channels. For example, if RGB channel, 
    padding should be tuple of size 3 like (0,0,0)
    '''
    if src.shape[0] > img_size[0] or src.shape[1] > img_size[1]:
        raise ValueError('new img_size has to be larger than the source size')
    
    new_img = np.full(img_size, padding, dtype = np.uint8)
    
    old_img_width, old_img_height, _ = src.shape
    new_img_width, new_img_height = img_size[0], img_size[1]
    
    x_center = (new_img_width - old_img_width) // 2
    y_center = (new_img_height - old_img_height) // 2
    
    new_img[x_center:x_center + old_img_width,y_center:y_center + old_img_height] = img
    
    return new_img
    


def plot_samples(X, y, labels, n = 20):
    '''
    Sample desired number of images from the specificed dataset to create a gridplot.
    
    '''
    # stratify datasets based on each category
    keys = list(labels.keys())
    
    X1 = X[y == keys[0]][:n]
    X2 = X[y == keys[1]][:n]
    
    cols = 10
    rows = int(n/cols)
           

    for i, imgs in enumerate([X1, X2]):

        plt.figure(figsize = (15, 6))

        for idx, img in enumerate(imgs):
            plt.subplot(rows, cols, idx+1)
            plt.imshow(img)
            plt.axis('off')
        plt.suptitle(f'Tumor: {labels[keys[i]]}')
        plt.show()
        

def grouped_bar_plot(datasets, class_name, width = 0.35,
                    labels = ['Training Set', 'Validation Set', 'Test Set'],
                    ylabel = 'Count', title = 'Count of classes in each set',
                    **kwargs):
    x = np.arange(len(labels))
    
    x1 = datasets[0]
    x2 = datasets[1]
    
    fig, ax = plt.subplots()
    ax.bar(x-width/2, x1, width, label = class_name[0])
    ax.bar(x+width/2, x2, width, label = class_name[1])
    
    ax.set_ylabel(ylabel)
    ax.set_title(title)
    ax.set_xticks(x, labels = labels)
    ax.grid(True)
    ax.legend()
    
    fig.tight_layout()
    plt.show()
```

### 3.3 Crop Black Border of Image
#### 3.3.1 Create crop function

```python
def crop_images(datasets, thresh_value = 30):
    '''
    Find the boundary box to crop black border of the image.
    cv2.findContour used to find the largest contour, and use it to find the boundary box
    
    thresh_value: it is a hyperparameter that will determine the final accuarcy of cropping
    
    '''
    new_datasets = []
    
    for img in datasets:
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        gray = cv2.GaussianBlur(gray, (5,5), 0)
        
        res, thresh = cv2.threshold(gray, thresh_value, 255, cv2.THRESH_BINARY)
        thresh = cv2.erode(thresh, None, iterations = 1)
        thresh = cv2.dilate(thresh, None, iterations = 2)
        
        # findContours works best on binary images
        cnts, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
        c = max(cnts, key = cv2.contourArea)
        
        # use the max contour to return the boundary rectangular box
        x, y, w, h = cv2.boundingRect(c)
        # crop the image to remove noninformative borders
        new_img = img[y:y+h, x:x+w, :]
        new_datasets.append(new_img)
        
    return np.array(new_datasets, dtype = object)
        
```

```python
X_train_croped = crop_images(X_train)
X_val_croped = crop_images(X_val)
X_test_croped = crop_images(X_test)
```

#### 3.3.2 Save cropped images

```python
def save_imgs(X, y, save_to_dir):
    '''
    X is the numpyarray of images
    y is the numpyarray of labels
    '''
    for i, (img, label) in tqdm(enumerate(zip(X, y)), total = len(y),
                               desc = 'Saving images: '):
        subfolder = os.path.join(save_to_dir, labels[label])
        Path(subfolder).mkdir(parents=True, exist_ok=True)
        
        dest = os.path.join(subfolder, str(i) + '.jpg')
        cv2.imwrite(dest, img)
```

```python
save_imgs(X_train_croped, y_train, './datasets/TRAIN_CROP')
save_imgs(X_test_croped, y_test, './datasets/TEST_CROP')
save_imgs(X_val_croped, y_val, './datasets/VAL_CROP')
```

## 4. Data Augmentation and Preprocessing
### 4.1 Create `ImageDataGenerator` 
VGG-16 preprocess_input does not come with scaling. So the rescale is included inside the `ImageDataGenerator`

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_generator = ImageDataGenerator(
                rescale = 1./255,
                rotation_range= 20,
                width_shift_range= 0.05,
                height_shift_range=0.05,
                shear_range=0.05,
                brightness_range=[0.1, 1.5],
                horizontal_flip=True,
                vertical_flip=True
                )
test_generator = ImageDataGenerator(
                rescale = 1./255)
```

Apply `ImageDataGenerator` to a sample data to see the augmentation and preprocessing

```python
IMG_SIZE = (150,150)

train_ds = train_generator.flow_from_directory('./datasets/TRAIN_CROP/',
                                              target_size = IMG_SIZE,
                                              batch_size=32,
                                              class_mode='binary')
valid_ds = test_generator.flow_from_directory('./datasets/VAL_CROP/',
                                             target_size = IMG_SIZE,
                                             batch_size= 16,
                                             class_mode='binary')
```

## 5. Model Building and Training
### 5.1 Pretrained VGG16 model

```python
from tensorflow.keras.applications.vgg16 import VGG16, preprocess_input
from tensorflow.keras import Model, Input
from tensorflow.keras.layers import Flatten, Dropout, Dense

NUM_CLASSES = 1

base_model = VGG16(include_top=False,weights = 'imagenet')
base_model.trainable = False

inputs = Input(shape = (150,150,3))
x = preprocess_input(inputs)
x = base_model(x)
x = Flatten()(x)
x = Dropout(0.2)(x)
outputs = Dense(NUM_CLASSES, activation='sigmoid')(x)

model = Model(inputs, outputs)

display(model.summary())
```

### 5.2 Fine-tune the model

```python
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping, CSVLogger
from tensorflow.keras import optimizers

NUM_EPOCHS = 50

model.compile(loss = 'binary_crossentropy',
             optimizer = optimizers.RMSprop(learning_rate=1e-4),
             metrics=['accuracy'])

fname = './checkpoints/vgg16_pretrained/noscaling002_CROPED'
Path(fname).mkdir(parents=True, exist_ok = True)

callbacks = [
      ModelCheckpoint(fname + '/pretrained.h5',
                      monitor = 'val_accuracy', save_best_only = True),
      CSVLogger(fname + '/pretrained_hist.csv',
                separator = ',', append = False),
]

history = model.fit(train_ds, epochs=NUM_EPOCHS,
                    validation_data=valid_ds,
                    callbacks = callbacks)
```

```python
def history_plot(history, **kwargs):
  
    plt.figure(figsize =(11,4))
    plt.subplot(121)
    plt.plot(history['accuracy'])
    plt.plot(history['val_accuracy'])
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.title('Model Accuracy')
    plt.legend(['Train', 'Validate'])

    plt.subplot(122)
    plt.plot(history['loss'])
    plt.plot(history['val_loss'])
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.title('Model Loss')
    plt.legend(['Train', 'Validate'])
    plt.tight_layout()
    plt.show()
```

## 6. Feature extraction using pretrained VGG16 network for other machine learning
### 6.1 Extract image features using pretrained VGG16 without its top layer

```python
from tensorflow.keras.applications.vgg16 import VGG16
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import regularizers

def bottlenect_features(data_dir, img_size = (150, 150),
                        batch_size = 8):
    datagen = ImageDataGenerator(rescale = 1./255)
  
    vgg16_model = VGG16(include_top = False, weights = 'imagenet')
    

    generator_train = datagen.flow_from_directory(data_dir,
                                        target_size = img_size,
                                        batch_size = batch_size,
                                        class_mode = None,
                                        shuffle = None)
    return vgg16_model.predict(generator_train)

def num_samples(data_dir):

    files = [f for f in os.listdir(data_dir) 
            if not f.startswith('.')]
    return len(files)

train_data = bottlenect_features('./datasets/train/')
num_train_yes = num_samples('./datasets/train/yes')
num_train_no = num_samples('./datasets/train/no')
train_labels = np.array([0]*num_train_no + [1]*num_train_yes)

val_data = bottlenect_features('./datasets/valid/')
num_val_yes = num_samples('./datasets/valid/yes')
num_val_no = num_samples('./datasets/valid/no')
val_labels = np.array([0]*num_val_no + [1]*num_val_yes)

```

### 6.2 Reshape the image data to suit it for traditional machine learning model

```python
from sklearn.utils import shuffle

X_train = train_data.reshape((train_data.shape[0], -1))
X_val = val_data.reshape((val_data.shape[0], -1))
y_train, y_val = train_labels, val_labels

# shuffle the datasets to permutate two samples arrangment in the datasets
X_train, y_train = shuffle(X_train, y_train, random_state=2022)
X_val, y_val = shuffle(X_val, y_val, random_state = 2022)
```

### 6.3 Preprocessing and normalize the data for further machine learning

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)

X_train_scaler = scaler.transform(X_train)
X_val_scaler = scaler.transform(X_val)
```

```python
from sklearn.linear_model import SGDClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier

#sgd_clf = SGDClassifier(random_state = 2022)
#sgd_clf.fit(X_train_scaler, y_train)

rf_clf = RandomForestClassifier()
rf_clf = rf_clf.fit(X_train_scaler, y_train)
```

### 6.3 Cross-validation to measure the model performance

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.base import clone

def diy_cross_val_score(clf, X, y, cv = 3, random_state = 2022):
    '''
    The function is customized to perform the same function as the
    cross_val_score provided in sklearn.
    '''
    skfolds = StratifiedKFold(n_splits=cv, shuffle = True,
                             random_state=random_state)
    
    score = []
    for train_index, test_index in skfolds.split(X, y):
        clone_clf = clone(clf)
        X_train_folds = X[train_index]
        y_train_folds = y[train_index]
        X_test_folds = X[test_index]
        y_test_folds = y[test_index]
        
        clone_clf.fit(X_train_folds, y_train_folds)
        y_pred = clone_clf.predict(X_test_folds)
        n_correct = sum(y_pred == y_test_folds)
        score.append(round(n_correct/len(y_pred), 4))
    return score
```

```python
from sklearn.model_selection import RandomizedSearchCV

n_estimators = [int(x) for x in np.linspace(200, 2000, 10)]
max_features = ['auto', 'sqrt']
max_depth = [int(x) for x in np.linspace(10,110,11)]
max_depth.append(None)
min_samples_split = [2,5,10]
min_samples_leaf = [1,2,4]
bootstrap = [True, False]

random_grid = [{'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}]

rf = RandomForestClassifier()
rf_clf_tuned = RandomizedSearchCV(estimator = rf,
            param_distributions=random_grid, n_iter=100,
            cv = 3, verbose = 2, random_state = 2022, 
                                  n_jobs =-1)
rf_clf_tuned.fit(X_train_scaler, y_train)
```

```python
best_params = rf_clf_tuned.best_params_
rf_clf_best = rf_clf_tuned.best_estimator_

best_params
```

```python
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay,accuracy_score

y_train_pred = cross_val_predict(rf_clf_best, X_train_scaler, y_train, cv = 5)
conf_mx = confusion_matrix(y_train, y_train_pred)
accuracy = accuracy_score(y_train, y_train_pred)
print(f'Accuracy of the Random Forest Classifier is: {accuracy*100:.2f}%')

display = ConfusionMatrixDisplay(conf_mx, display_labels=['No', 'Yes']).plot()
plt.show()
```
